```python
import numpy as np
import pandas as pd
import yfinance as yf
```

```python
spy_ohlc_df = yf.download('SPY', start='1990-01-01', end='2022-03-01')
```

```python
spy_ohlc_df.index.name = 'date'
spy_ohlc_df = spy_ohlc_df.rename(columns = {'Open': 'open', 'High':'high', 'Low':'low', 'Close':'close', 'Adj Close': 'adj_close', 'Volume': 'volume'})
```

```python
spy_ohlc_df['res'] = spy_ohlc_df['close']/spy_ohlc_df['close'].shift(1)
spy_ohlc_df['res_log'] = np.log(spy_ohlc_df['res'])
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(40,10))
plt.yscale('log')
plt.plot(spy_ohlc_df.index, spy_ohlc_df['close'])
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(40,10))
plt.yscale('log')
plt.plot(spy_ohlc_df.index[-1000:], spy_ohlc_df['close'][-1000:])

import matplotlib.pyplot as plt
plt.figure(figsize=(40,10))
# plt.yscale('log')
plt.plot(spy_ohlc_df.index[-1000:], spy_ohlc_df['res_log'][-1000:])
plt.plot([spy_ohlc_df.index[-1000:].min(), spy_ohlc_df.index[-1000:].max()], [0,0])
```

```python
plt.figure(figsize=(40,10))
plt.xticks(ticks=np.arange(-0.15, 0.15, 0.005))
plt.hist(spy_ohlc_df['res_log'], bins = 500, alpha = 0.5)
plt.hist(-spy_ohlc_df['res_log'], bins = 500, alpha = 0.5)
plt.plot()
plt.show()
```

```python
i=365
spy_ohlc_df['close_prev'] = spy_ohlc_df['close'].shift(i)
spy_ohlc_df['profit'] = spy_ohlc_df['close']/spy_ohlc_df['close_prev']
spy_ohlc_df['profit_prev'] = spy_ohlc_df['profit'].shift(i)

print(spy_ohlc_df.mean())
```
